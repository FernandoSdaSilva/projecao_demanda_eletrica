### Estacionariedade

```python
#https://www.statology.org/kpss-test-in-python/
import statsmodels.api as sm

def print_kpss(data):
    test_stat, p_value, lag_trunc, critical_values = sm.tsa.stattools.kpss(data, regression='ct')
    print(f">Estatística de teste: {test_stat}\n>p-valor: {p_value}\n>Truncation Lag Parameter: {lag_trunc}\n>Valores críticos: {critical_values}")
```

```python
def kpss_test(series, **kw):    
    statistic, p_value, n_lags, critical_values = kpss(series, **kw)
    # Format Output
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    print(f'Result: The series is {"not " if p_value < 0.05 else ""}stationary')
```

```python
kpss_test(df_load_sul["load_mwmed"])
```

```python
print_kpss(df_load_sul["load_mwmed"])
```
