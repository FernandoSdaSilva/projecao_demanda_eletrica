```python
df_poa = df[df["ESTACAO"] == "PORTO ALEGRE"]
```

```python
df_poa.head()
```

```python
df_poa["TEMPERATURA_DO_AR__BULBO_SECO_HORARIA"].describe()  # contém -9999 para valores vazios (aparentemente)
```

```python
df_poa.replace(-9999, 0, inplace = True)    # substituindo -9999 com 0
```

```python
plt.figure(figsize = (20, 5))

sns.lineplot(x = "DATA_HORA", y = "TEMPERATURA_DO_AR__BULBO_SECO_HORARIA", data = df_poa, hue = df_poa["DATA_HORA"].dt.year.astype("str"))
#sns.lineplot(x = "DATA_HORA", y = "TEMPERATURA_MÁXIMA_NA_HORA_ANT", data = df_poa, color = "gray")
#sns.lineplot(x = "DATA_HORA", y = "TEMPERATURA_MÍNIMA_NA_HORA_ANT", data = df_poa, color = "gray")


plt.show()
```

```python
plt.figure(figsize = (20, 5))

sns.lineplot(x = "DATA_HORA", y = "PRECIPITAÇÃO_TOTAL_HORÁRIO", data = df_poa)

plt.show()
```

```python
df_poa_daily = df_poa[["DATA", "TEMPERATURA_DO_AR__BULBO_SECO_HORARIA", "PRECIPITAÇÃO_TOTAL_HORÁRIO"]].groupby("DATA", as_index = False).mean()
df_poa_daily["ANO"] = df_poa_daily["DATA"].dt.year.astype("str")
df_poa_daily.head()
```

```python
plt.figure(figsize = (20, 5))

sns.lineplot(x = "DATA", y = "TEMPERATURA_DO_AR__BULBO_SECO_HORARIA", data = df_poa_daily, hue = "ANO", palette = "magma")

plt.show()
```

```python
df_poa["ANO_MES"] = pd.to_datetime(df_poa["DATA"].dt.year.astype("str") + df_poa["DATA"].dt.month.astype("str") + "01", format = "%Y%m%d")
df_poa_monthly = df_poa[["ANO_MES", "TEMPERATURA_DO_AR__BULBO_SECO_HORARIA", "PRECIPITAÇÃO_TOTAL_HORÁRIO"]].groupby("ANO_MES", as_index = False).mean()
df_poa_monthly["ANO"] = df_poa_monthly["ANO_MES"].dt.year.astype("str")
#df_poa_monthly["ANO"] = df_poa_monthly["DATA"].dt.year.astype("str")
df_poa_monthly.head()
```

```python
plt.figure(figsize = (20, 5))

sns.lineplot(x = "ANO_MES", y = "TEMPERATURA_DO_AR__BULBO_SECO_HORARIA", data = df_poa_monthly, hue = "ANO", palette = "magma")

plt.show()
```
